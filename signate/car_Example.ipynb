```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
```

```python
data=pd.read_csv("train.tsv",delimiter="\t")
```

```python
data
```

```python
#データを入力データと正解データに分離
train_x = data.values[:,2:9]
train_y = data.values[:,1]
```

```python
#モデルの作成
model = RandomForestRegressor(n_estimators=100)
```

```python
#モデルの学習(エラー発生)
model.fit(train_x,train_y)
```

```python
#データに"?"があるためエラーが発生
train_x[:,2]
```

```python
#"?"にはデータが存在しないため、全体の平均値で置換

power = []
for i in range(len(train_x)):
    if train_x[i,2]!='?':
        power.append(float(train_x[i,2]))

#平均値の算出
mean_power=np.mean(power)

for i in range(len(train_x)):
    if train_x[i,2]=='?':
        train_x[i,2]=mean_power
    else:
        train_x[i,2]=float(train_x[i,2])
```

```python
#2列目の要素をNoneとfloatに変換した結果
train_x[:,2]
```

```python
#モデルの学習
model.fit(train_x,train_y)
```

```python
#予測データの作成
test=pd.read_csv("test.tsv",delimiter="\t")
test_x = test.values[:,1:8]
```

```python
for i in range(len(test_x)):
    if test_x[i,2]=='?':
        test_x[i,2]=mean_power
    else:
        test_x[i,2]=float(train_x[i,2])
```

```python
#学習したモデルで予測
model.predict(test_x)
```
